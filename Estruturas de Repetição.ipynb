#### 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

````python
# 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.
nota = input("Informe a nota: ")
while not nota.isdigit() or int(nota) < 0 or int(nota) > 10:
    print("Valor inválido. A nota deve ser um número entre 0 e 10.")
    nota = input("Informe a nota: ")
print(f"A nota informada é: {nota}")
````

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

#### 4. Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

````python
paisA = 80000
paisB = 200000
anos = 0

while paisB > paisA:
    paisA *= 1.03
    paisB *= 1.015
    anos += 1

print("Em {} anos o País B terá uma população maior que o País A" .format(anos))
````

#### 5. Altere o programa anterior permitindo ao usuário informar as populações e as taxas de crescimento iniciais. Valide a entrada e permita repetir a operação.

#### 8. Faça um programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa); e então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

````python
qtd = int(input("Informe a quantidade de pessoas: "))
total = 0 

for i in range (qtd):
    idade = int(input("Informe a idade da {}ª pessoa: " .format(i+1)))
    total += idade

media = total/qtd

if media >= 0 and media <=25:
    print("Grupo Jovem -> Média das {} pessoas: {}" .format(qtd, media))
elif media >= 26 and media <= 60:
    print("Grupo Sênior -> Média das {} pessoas: {}" .format(qtd, media))
else: 
    print("Grupo idoso -> Média das {} pessoas: {}" .format(qtd, media))
````

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

````python
qtd_eleitores = int(input("Quantas pessoas irão votar? "))
candidato1 = 0
candidato2 = 0
candidato3 = 0

for i in range (qtd_eleitores):
    voto = int(input("Informe o número do candidato: "))
    if voto == 1:
        candidato1 += 1
    elif voto == 2:
        candidato2 += 1
    elif voto == 3:
        candidato3 += 1

print("""
      Candidato 1 -> {} votos
      Candidato 2 -> {} votos
      Candidato 3 -> {} votos""" .format(candidato1, candidato2, candidato3))
````

#### 10. Faça um programa que calcule o valor total investido por um colecionador em sua coleção de CDs e o valor médio gasto em cada um deles. O usuário deverá informar a quantidade de CDs e o valor para em cada um.

````python
qtd = int(input("Informe a quantidade de discos na sua coleção: "))
total_gasto = 0

print ("Informe o valor de cada um dos {} discos: " .format(qtd))
for i in range(qtd): 
    valor = float(input("Valor do {} disco: " .format(i+1)))
    total_gasto += valor

media = total_gasto / qtd

print ("Média de valores gastos com os discos: {}" .format(media))

````

#### 12. Um funcionário de uma empresa recebe aumento salarial anualmente: Sabe-se que:
<pre>
Esse funcionário foi contratado em 1995, com salário inicial de R$ 1.000,00;
Em 1996 recebeu aumento de 1,5% sobre seu salário inicial;
A partir de 1997 (inclusive), os aumentos salariais sempre correspondem ao dobro do percentual do ano anterior. Faça um programa que determine o salário desse funcionário em 2003. 
</pre>

````python
salario = 1000 * 1.015
porcentagem = 0.015

for i in range(1997, 2004):
    porcentagem *= 2
    salario *= (1 + porcentagem)

print('O salário em 2003 era R$ {:.2f}'.format(salario))
````

````python
total = 0
codigo = input('Informe o código do produto, caso deseje encerrar digite "encerrar": ')
while codigo != 'encerrar':
    qtde = int(input('Informe a quantidade: '))
    if codigo == '100':
        total += 1.2 * qtde
    elif codigo == '101':
        total += 1.3 * qtde
    elif codigo == '102':
        total += 1.5 * qtde
    elif codigo == '103':
        total += 1.2 * qtde
    elif codigo == '104':
        total += 1.3 * qtde
    elif codigo == '105':
        total += 1 * qtde
        
    codigo = input('Informe o código do produto, caso deseje encerrar digite "encerrar": ')

print('O total a ser pago é: R$ ', total)
````

```python
# Existem formas diferentes de importar um biblioteca

# import biblioteca -> Dessa maneira toda vez que for usar alguma ferramenta da biblioteca é necessário nomeá-la. Exemplo: print(time.ctime())

#from biblioteca import * -> Importa todas as ferramentas/métodos de um biblioteca. Não precisa nomeá-la. Exemplo: print(ctime())

#from biblioteca import ctime -> Importa uma ferramenta específica da bilioteca.
```

```python
# Falando sobre a biblioteca Time

import time

# Marco zero: EPOCH (01 de janeiro de 1970 as 00:00:00)
# time() retorna quantos segundos se passaram desde a EPOCH
print(time.time())


# Isso pode ser muito útil para medir o tempo que um código leva para rodar

tempo_inicial = time.time()

for i in range (100000000):
    pass

tempo_final = time.time()

print(f"meu código demorou {tempo_final - tempo_inicial} para rodar")

# ctime() retorna a data e hora atual em string

print(time.ctime())
```

```python
# Fazer o progama esperar um tempo até o próximo comando time.sleep(). Muito útil em automações

print("Abrir o sistema")

time.sleep(5) # espera 5 segundo até a próxima ação

print("Entrar no sistema")
```

```python
# gmtime()
# localtime()
#Devolvem a hora em formato de dicionário, me permitindo pegar uma informação esécífica dentro da data
data_global = time.gmtime() #segue o fuso horario padrao greenwich
data_local = time.localtime()

print(data_global)
print(data_local)

dia_hoje = data_local.tm_mday
mes_hoje = data_local.tm_mon
ano_hoje = data_local.tm_year
print(f"Dat atual: {dia_hoje}/{mes_hoje}/{ano_hoje}")
```

```python
data_texto = time.strftime("%d-%m-%Y || %A", data_local) #Transforma a data em uma string. O codigo de formatação varia e está disponível online para consulta
print(data_texto)
```

```python
# Se quiser trazer traduzido...
# É necessário informar para o python o seu local, usando a biblioteca locale

import locale
locale.setlocale(locale.LC_TIME, "pt_BR")# Na documentação da biblioteca encontramos a forma de configurar 

data_texto = time.strftime("%A, %d de %B de %Y", data_local) 
print(data_texto)
```

```python
# Para ler arquivos com datas formatadas de maneira diferente 
lista_datas = ["01 de Junho de 2026", "01 de Janeiro de 2027", "15 de Fevereiro de 2027", "18 de Dezembro de 2027"]

for data in lista_datas:
    data_estruturada = time.strptime(data, "%d de %B de %Y")
    print(data_estruturada)
```

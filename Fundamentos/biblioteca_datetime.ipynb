```python
#datetime.datetime.now()

from datetime import datetime
data_hora_atual = datetime.now()
print(data_hora_atual)
```

```python
#.date()
#.time()

data = data_hora_atual.date()
print(data)

hora = data_hora_atual.time()
print(hora)

ano = data_hora_atual.year
print(ano)
mes = data_hora_atual.month
print(mes)
dia = data_hora_atual.day
print(dia)
hora = data_hora_atual.hour
print(hora)
minutos = data_hora_atual.minute
print(minutos)
segundos = data_hora_atual.second
print(segundos)
```

```python
# função today()

from datetime import date
dia_hoje = date.today()
print(dia_hoje)

ano = dia_hoje.year
print(ano)
mes = dia_hoje.month
print(mes)
dia = dia_hoje.day
print(dia)
```

```python
# Para fazer operações com data
from datetime import datetime, timedelta

data_inicio = datetime(2026,6,25)
duracao = 387

data_final = data_inicio + timedelta(days=duracao)
print(data_final)

# se fosse em semanas?

data_final = data_inicio + timedelta(weeks=duracao)
print(data_final)

# se fosse em minutos?

data_final = data_inicio + timedelta(minutes=duracao)
print(data_final)


# Fazendo o processo contrario

tempo_projeto = data_final - data_inicio
print(tempo_projeto.)
print(type(tempo_projeto))

```

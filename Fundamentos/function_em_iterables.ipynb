```python
# lista = list(map(função, terable_original))

def padronizar_texto (texto): # Função para padronizar os textos de uma lista
    texto = texto.casefold()
    texto = texto.replace(" ", "")
    texto = texto.strip()
    return texto


produtos = ['ABc12', 'abc34', 'AbC37', 'beb12', 'BSA151', 'BeB23'] #Lista de produtos
```

```python
# Fazendo por for usando a function

for i, produto in enumerate (produtos):
    produtos[i] = padronizar_texto(produto)
print(produtos)
```

```python
# Fazendo com list map usando a function

produtos = list(map(padronizar_texto, produtos))
print(produtos)
```

```python
# Aplicando function sort

# o sort ordena baseado na tabela ASCII onde são atribuídos valores numericos a letras.
#Nessa tabela, letras maiusculas sempre vem na frente de letras minusculas. Por isso nem sempre a ordem alfabetica fica correta
produtos.sort() 
print(produtos)
```

```python
# Para resolver esse problema, podemos passar uma 'condição' para o sort

produtos.sort(key = str.casefold) #Nesse caso, o .sort considera todas as letras como se fossem minusculas e só então faz a ordenação dos itens.

print(produtos)
#Como vemos no resultado a function .sort não altera os valores na lista. Ele apenas considera como minuscula para ordenação.
```

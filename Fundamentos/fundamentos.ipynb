### Estrutura While:
Usado quando queremos repetir um código de forma indeterminada. Enquanto uma condição for verdadeira/falsa


 while condição:
    repete o código

```python
venda = input ("Registre um produto. Para cancelar o registro, basta apertar enter com a caixa vazia")
vendas = []

while venda != "":
    vendas.append(venda)
    venda = input ("Registre um produto> Para cancelar o registro, basta apertar enter com a caixa vazia")
    
print ("Registro finalizado. Vendas cadastradas: {}" .format(vendas))
```

## Dicionários em python

#### Estrutura: 
dicionario = {chave: valor, chave: valor, chave: valor...}

```python
mais_vendidos = {'tecnologia': 'iphone', 'refrigeracao': 'Ar consul 120000btu', 'livro': 'O Alquimista', 'eletrodomestico': 'Cafeteira Dolce Gusto'}

vendas_tecnologia = {'iphone': 15000, 'samsung galaxy': 12000, 'TV LG 50"': 10000, 'Xbox Series S': 56000}

#Qual o item mais vendido em eletrodmoestico?
eletrodomestico = mais_vendidos['eletrodomestico']
print(eletrodomestico)
#ou
print(mais_vendidos['eletrodomestico'])

# Qual o item mais vendido em livro?
livro = mais_vendidos['livro']
print(livro)
#ou
print(mais_vendidos['livro'])

# Quantos Xbox foram vendidos?
xbox = vendas_tecnologia['Xbox Series S']
print(xbox)
#ou
print(vendas_tecnologia['Xbox Series S'])

# Quantas Tv's foram vendidas?
tv = vendas_tecnologia['TV LG 50"']
print(tv)
#ou
print(vendas_tecnologia['TV LG 50"'])
```

```python
#usando o metodo get

#Qual o item mais vendido em refrigeração?
print(mais_vendidos.get('refrigeracao'))

#Qual o item mais vendido em tecnologia?
print(mais_vendidos.get('tecnologia'))

#Quantos Galaxy foram vendidos?
print(vendas_tecnologia.get('samsung galaxy'))

#Quantos Iphone formam vendidos?
print(vendas_tecnologia.get('iphone'))
```

```python
#Verificando se um item está no dicionário if ou .get(chave) = None

#Ferramentas está listada entre os itens mais vendidos?
if 'ferramenta' in mais_vendidos:
    print(mais_vendidos['ferramenta'])
else: 
    print("Não há uma categaoria 'Ferramentas' entre os itens mais vendidos")

#usando get
if mais_vendidos.get('ferramenta') == None:
    print("Não há uma categaoria 'Ferramentas' entre os itens mais vendidos")
else:
    print(mais_vendidos.get('ferramenta'))

```

```python
mais_vendidos2 = {'lazer': 'Trionda', 'vestuario': 'Camisa 2 BOTAFOGO 2026', 'calcado':'Vans All Black'}
# Adicionar itens em um dicionário
#dicionario[chave] = valor ou dicionario.pudate({chave: valor, chave: valor...})

mais_vendidos['ferramenta'] = 'Parafusadeira Bosch 18V'
print(mais_vendidos['ferramenta'])

#Adicionando vários itens ou um dicionario no outro
mais_vendidos.update(mais_vendidos2)
print(mais_vendidos)

# modificando itens. É só "adicionar" um item que já existe
mais_vendidos['lazer'] = 'Barraca de camping'
print(mais_vendidos['lazer'])


#Removendo um item 
# del dicionario[chave] -> O método del pode ser usado também em uma lista - del lista[0]
#Ou
#valor = dicionario.pop(chave) -> o método pop possibilita armazenar o item retirado em uma variavel 
del mais_vendidos['calcado']
print(mais_vendidos)

#clear limpa o dicionario
mais_vendidos2.clear
print(mais_vendidos2)
```

```python
# Usando o for dentro de um dicionário
# for chave in dicionario:
    #faça alguma coisa

for chave in vendas_tecnologia:
    print('{}: {} unidades' .format(chave, vendas_tecnologia[chave]))
```

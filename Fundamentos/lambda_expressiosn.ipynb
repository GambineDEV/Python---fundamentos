## Lambda Expressions

#### Objetivo:

* As Lambdas expressios são funções anônimas (sem nome) que tem   linha de código e são atribuídas a uma variável, como se a variável virasse uma função.

* Elas normalmente são usadas para fazer uma única ação. Em python, geralmente são usadas dentro de métodos como argumento, para não precisarmos criar uma função só para isso.

* Outra aplicação está em criar um "gerador de funções"


#### Estrutura:
minha_funcao = lambda parametro : expressao

```python
# Criando uma função que executa apenas 1 linha de código
def minha_funcao (num):
    return num * 2
print(minha_funcao(5))




# Usando lambda expression
minha_funcao2 = lambda num: num*2

print(minha_funcao2(5))
```

```python
# Mais um exemplo: Criar uma função que calcula o preço dos produtos acrescido do imposto

imposto = 0.3


def preco_imposto(preco):
    return preco *(1 + 0.3)
print(preco_imposto(100))


preco_imposto2 = lambda preco: preco*(1 + imposto)
print(preco_imposto2(100))
```

```python
# aplicando lambda como argumento de alguma outra função / map e filter

preco_tecnologia = {'notebook asus' : 2450, 'iphone': 4500, 'samsung galaxy': 3000, ' tv samsung': 1000, 'ps5': 3000, 'tablet':3000, 'smartwatch' : 1250, 'notebook dell': 3600, 'ipad': 2750}
```

```python
# map() -> Queremos saber o preço de cada produto adicionando o valor do imposto de 30% sobre o valor do produto

# Fazendo por function
def calcular_preco (preco):
    return preco * 1.3

preco_com_imposto = list(map(calcular_preco, preco_tecnologia.values()))
print(preco_com_imposto) 
```

```python
# fazendo por lambda expression

preco_com_imposto2 = list(map(lambda preco: preco*1.3, preco_tecnologia.values()))
print(preco_com_imposto2)
```

```python
# Queremos apenas os produtos que custam acima de 2000
# filter(função, iterable) -> retorna como resposta todos os itens do iterable onde a função é True


# fazendo por function
def maior_que2000 (item):
    return item[1] > 2000

produtos_acima2000 = dict(list(filter(maior_que2000, preco_tecnologia.items())))
print(produtos_acima2000)
```

```python
# fazendo por lambda

produtos2_acima2000 = dict(list(filter(lambda item: item[1] > 2000, preco_tecnologia.items())))
print(produtos2_acima2000)
```

## Lambda expression para gerar funções

#### Descrição

* Uma das aplicações de lambda expressions é criar um 'gerador de funções'. Nesse caso, usaremos a lambda expressions dentro da definição de uma outra função

#### Exemplo: 

1. Vamos criar uma função que me permita calcualar o valor acrescido do imposto de diferentes categorias (produto , serviço, royalties, etc.)

```python
# imposto de produto = 0.1
# imposto de serviço = 0.15
# imposto de royalties = 0.25

# Fazendo por funcoes

def calcular_preco_produto(preco):
    return preco*0.1

def calcula_preco_servico(preco):
    return preco*0.15

def calcula_preco_royaltie (preco):
    return preco*0.25
```

```python
# fazendo por lambda expression

def calcular_imposto(imposto):
    return lambda preco: preco*(1 + imposto)
```

```python
calcular_preco_produto = calcular_imposto(0.1)
calcular_preco_servico = calcular_imposto(0.15)
calcular_preco_royaltie = calcular_imposto(0.25)


print(int(calcular_preco_produto(100)))
print(int(calcular_preco_servico(100)))
print(int(calcular_preco_royaltie(100)))

```

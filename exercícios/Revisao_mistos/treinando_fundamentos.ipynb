### Exercício 3: Mapeamento de Vagas Freelance (List Comprehension)
Você extraiu dados de plataformas de trabalho remoto. Use **List Comprehension** para criar uma lista apenas com os *títulos* das vagas que exigem a habilidade `'Python'`.

```python
vagas = [
    {'titulo': 'Dev Backend', 'habilidades': ['Python', 'SQL', 'Docker']},
    {'titulo': 'Dev Frontend', 'habilidades': ['React', 'JavaScript', 'CSS']},
    {'titulo': 'Engenheiro de Dados', 'habilidades': ['Python', 'Spark', 'AWS']},
    {'titulo': 'Web Designer', 'habilidades': ['Figma', 'Photoshop']}
]

# TODO: Crie a lista 'vagas_python' usando list comprehension
vagas_python = [(vaga['titulo'], vaga['habilidades']) for vaga in vagas if 'Python' in vaga['habilidades']]
print(vagas_python)
```

```python
logs_rede = [
    ('192.168.0.10', 80, 'allowed'),
    ('10.0.0.5', 22, 'blocked'),
    ('192.168.0.11', 80, 'blocked'),
    ('172.16.0.2', 443, 'allowed'),
    ('10.0.0.8', 80, 'blocked')
]

# TODO: Percorra a lista e conte os bloqueios na porta 80
bloqueados_porta_80 = 0 
for ip, porta, status in logs_rede:
    if porta == 80 and status == 'blocked':
        bloqueados_porta_80 += 1
print("Quantidade de acessos bloqueados: {} " .format(bloqueios_porta_80))
```

### Exercício 6: Normalização de Dados de Sensores (Listas Aninhadas)
Na automação industrial, um sensor enviou leituras agrupadas em sublistas. Achate (flatten) essa matriz para uma lista simples contendo apenas as leituras maiores que `50`.

```python
matriz_leituras = [
    [20, 55, 10],
    [60, 45, 80],
    [90, 15, 30]
]

# TODO: Transforme matriz_leituras em uma lista unidimensional apenas com valores > 50
#usando for aninhado
lista = []
for linha in matriz_leituras:
    for leitura in linha:
        if leitura > 50:
            lista.append(leitura)
print(f"Leituras acima de 50: {lista}")

#ou

#usando list comprehension com for aninhado

lista_filtrada =  [leitura for linha in matriz_leituras for leitura in linha if leitura > 50 ]
print (f"Leituras acima de 50: {lista_filtrada}")
```

### Exercício 8: Atualização de Estoque em Lote (Iteração e Modificação)
Você tem o estoque atual do armazém e uma lista de tuplas com os despachos do dia `(id_produto, quantidade_despachada)`. Atualize o dicionário de estoque subtraindo os despachos.

```python
estoque = {'A1': 100, 'B2': 50, 'C3': 200}
despachos_hoje = [('A1', 20), ('C3', 150), ('B2', 50)]

# TODO: Atualize o dicionário 'estoque'

for produto, qtd in despachos_hoje:
    if produto in estoque:
        estoque[produto] -= qtd
print(estoque) # Esperado: {'A1': 80, 'B2': 0, 'C3': 50}
```

### Exercício 9: Loop Infinito Seguro (While True e Break)
Crie a lógica de um menu de aplicativo de treinos físicos. Use `while True:` com a função `input()` para pedir um comando. Se o usuário digitar `'1'`, imprima 'Registrando treino'. Se digitar `'sair'`, use `break` para encerrar o loop.

```python
# TODO: Implemente o while True com input e break
while True:
    op = input("Digite 1 para registrar um treino ou 'sair' para encerrar: ")
    if op == '1':
        print("Registrando treino")
    elif op.lower() == 'sair':
        print("Encerrando")
        break
    else:
        print("Comando inválido!")
```

## Oferecendo desconto para cliente com base na última compra

Suponha que você está trabalhando para uma empresa que deseja rastrear a atividade do cliente. Uma métrica que eles estão interessados é o tempo que passou desde a última transação do cliente. Se for muito tempo, eles podem oferecer um desconto para o cliente. Crie um script Python que mostra quanto tempo se passou desde a última compra do cliente. Se faz mais de 30 dias, mostre uma mensagem oferecendo um desconto para o cliente.

```python
from datetime import datetime

ultima_compra = datetime (2026, 1, 10)
data_hora_atual = datetime.now()

print(ultima_compra)
print(data_hora_atual)


diferenca = data_hora_atual - ultima_compra
if diferenca.days > 30:
    print("Parabéns, você ganohu um desconto!")
```

## Data e hora em diferentes fusos horários

Uma empresa tem escritórios em São Paulo, Nova York e Tóquio. Crie um script Python que mostra a data e hora atuais nesses três fusos horários. Exiba, também, se estes escritórios estão abertos ou fechados (9h às 17h).


```python
from datetime import datetime
from zoneinfo import ZoneInfo

data_hora_atual = datetime.now()
fuso_sp = ZoneInfo("America/Sao_Paulo")
fuso_ny = ZoneInfo("America/New_York")
fuso_toquio = ZoneInfo("Asia/Tokyo")

data_hora_sp = data_hora_atual.astimezone(fuso_sp)
data_hora_ny = data_hora_atual.astimezone(fuso_ny)
data_hora_toquio = data_hora_atual.astimezone(fuso_toquio)

print(f"Data e hora em São Paulo: {data_hora_sp}")
print(f"Data e hora em Nova York: {data_hora_ny}")
print(f"Data e hora em Tóquio: {data_hora_toquio}")


def verifica_horario (data_hora):
    if 9 <=data_hora.hour < 17:
        return "Escritório aberto"
    else:
        return"Escritório Fechado"



print(f"São Paulo: {verifica_horario(data_hora_sp)}")
print(f"Nova York: {verifica_horario(data_hora_ny)}")
print(f"Tóquio: {verifica_horario(data_hora_toquio)}")
```

## Calculando a idade

Um usuário fornece sua data de nascimento no formato "dd/mm/aaaa". Crie um Script que calcula a idade do usuário

```python
from datetime import datetime

data_nascimento = input("Informe sua data de nascimento(dd/mm/aaaa): ")
data_nascimento = datetime.strptime(data_nascimento, "%d/%m/%Y")
data_atual = datetime.now()

idade = data_atual.year - data_nascimento.year

if data_nascimento.month > data_atual.month:
    print(f"O usuário tem {idade - 1} anos")
elif data_nascimento.month < data_atual.month:
    print(f"O usuário tem {idade} anos")
elif data_nascimento.month == data_atual.month:
    if data_nascimento.day > data_atual.day:
        print(f"O usuário tem {idade - 1} anos")
    elif data_nascimento.day < data_atual.day:
        print(f"O usuário tem {idade} anos")
    else:
        print(f"O usuário completa {idade} anos hoje!")
```

### Estrutura While:
Usado quando queremos repetir um código de forma indeterminada. Enquanto uma condição for verdadeira/falsa


 while condição:
    repete o código

```python
venda = input ("Registre um produto. Para cancelar o registro, basta apertar enter com a caixa vazia")
vendas = []

while venda != "":
    vendas.append(venda)
    venda = input ("Registre um produto> Para cancelar o registro, basta apertar enter com a caixa vazia")
    
print ("Registro finalizado. Vendas cadastradas: {}" .format(vendas))
```
